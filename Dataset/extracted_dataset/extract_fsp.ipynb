Segregate into multi-segment and non-multisegment

```python
import os
import shutil

def segregate_fsp_files(folder_path):
    """
    Scans the given folder for .fsp files and moves them into two subfolders:
      - 'multisegment': for files containing "MULTISEGMENT MODEL "
      - 'non-multisegment': for files not containing the string
    """
    # Define paths for the target subfolders
    multiseg_folder = os.path.join(folder_path, 'multisegment')
    non_multiseg_folder = os.path.join(folder_path, 'non-multisegment')
    
    # Create the subfolders if they don't already exist
    os.makedirs(multiseg_folder, exist_ok=True)
    os.makedirs(non_multiseg_folder, exist_ok=True)
    
    # Iterate through all files in the folder
    for filename in os.listdir(folder_path):
        if filename.lower().endswith('.fsp'):
            file_path = os.path.join(folder_path, filename)
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    content = f.read()
                # Check if the file contains the target string
                if "MULTISEGMENT MODEL " in content:
                    target_path = os.path.join(multiseg_folder, filename)
                    shutil.move(file_path, target_path)
                    print(f"Moved '{filename}' to 'multisegment'")
                else:
                    target_path = os.path.join(non_multiseg_folder, filename)
                    shutil.move(file_path, target_path)
                    print(f"Moved '{filename}' to 'non-multisegment'")
            except Exception as e:
                print(f"Error processing {filename}: {e}")

if __name__ == '__main__':
    # Replace this with the path to your folder containing .fsp files
    folder_path = 'E:\SDA_CE\srcmod2024-12-02FSP'
    segregate_fsp_files(folder_path)
```

```python
import os
import glob
import re
import csv

def get_fieldnames(data_list, fixed_columns):
    """
    Given a list of dictionaries (data_list) and a list of fixed_columns,
    returns a list of field names that is the union of all keys:
      - fixed_columns appear first,
      - any additional keys (if present in some rows) are appended (sorted).
    """
    union_keys = set()
    for row in data_list:
        union_keys |= set(row.keys())
    additional = sorted(list(union_keys - set(fixed_columns)))
    return fixed_columns + additional

def parse_finite_source(block):
    """
    Parses the finite-source rupture model section.
    Expected values:
      Event, EventTAG, LAT, LON, DEP, LEN, WID, Mw, Mo,
      STRK, DIP, RAKE, Htop, HypX, HypZ, avTr, avVr
    """
    data = {}
    # Remove any leading '%' and spaces; split into lines
    lines = [line.strip('% ').strip() for line in block.splitlines() if line.strip()]
    for line in lines:
        if line.startswith("Event :"):
            # Example: "Event : Izu-hanto-oki (Japan)         05/09/1974         [Takeo (1990) ]"
            m = re.search(r"Event\s*:\s*([^\[]+)", line)
            if m:
                event = m.group(1).strip()
                # Remove any date in the string (format dd/mm/yyyy)
                event = re.sub(r"\d{1,2}/\d{1,2}/\d{4}", "", event).strip()
                data["Event"] = event
        elif line.startswith("EventTAG:"):
            m = re.search(r"EventTAG:\s*(\S+)", line)
            if m:
                data["EventTAG"] = m.group(1)
        elif line.startswith("Loc  :"):
            m_lat = re.search(r"LAT\s*=\s*([\d\.\-]+)", line)
            m_lon = re.search(r"LON\s*=\s*([\d\.\-]+)", line)
            m_dep = re.search(r"DEP\s*=\s*([\d\.\-]+)", line)
            if m_lat:
                data["LAT"] = float(m_lat.group(1))
            if m_lon:
                data["LON"] = float(m_lon.group(1))
            if m_dep:
                data["DEP"] = float(m_dep.group(1))
        elif line.startswith("Size :"):
            m_len = re.search(r"LEN\s*=\s*([\d\.Ee\+\-]+)", line)
            m_wid = re.search(r"WID\s*=\s*([\d\.Ee\+\-]+)", line)
            m_mw = re.search(r"Mw\s*=\s*([\d\.Ee\+\-]+)", line)
            m_mo = re.search(r"Mo\s*=\s*([\d\.Ee\+\-]+)", line)
            if m_len:
                data["LEN"] = float(m_len.group(1))
            if m_wid:
                data["WID"] = float(m_wid.group(1))
            if m_mw:
                data["Mw"] = float(m_mw.group(1))
            if m_mo:
                data["Mo"] = float(m_mo.group(1))
        elif line.startswith("Mech :"):
            m_strk = re.search(r"STRK\s*=\s*([\d\.Ee\+\-]+)", line)
            m_dip = re.search(r"DIP\s*=\s*([\d\.Ee\+\-]+)", line)
            m_rake = re.search(r"RAKE\s*=\s*([\d\.Ee\+\-]+)", line)
            m_htop = re.search(r"Htop\s*=\s*([\d\.Ee\+\-]+)", line)
            if m_strk:
                data["STRK"] = float(m_strk.group(1))
            if m_dip:
                data["DIP"] = float(m_dip.group(1))
            if m_rake:
                data["RAKE"] = float(m_rake.group(1))
            if m_htop:
                data["Htop"] = float(m_htop.group(1))
        elif line.startswith("Rupt :"):
            m_hypx = re.search(r"HypX\s*=\s*([\d\.Ee\+\-]+)", line)
            m_hypz = re.search(r"HypZ\s*=\s*([\d\.Ee\+\-]+)", line)
            m_avtr = re.search(r"avTr\s*=\s*([\d\.Ee\+\-]+)", line)
            m_avvr = re.search(r"avVr\s*=\s*([\d\.Ee\+\-]+)", line)
            if m_hypx:
                data["HypX"] = float(m_hypx.group(1))
            if m_hypz:
                data["HypZ"] = float(m_hypz.group(1))
            if m_avtr:
                data["avTr"] = float(m_avtr.group(1))
            if m_avvr:
                data["avVr"] = float(m_avvr.group(1))
    return data

def parse_inversion(block):
    """
    Parses the inversion-related parameters section.
    Expected values include:
      Nx, Nz, Fmin, Fmax, Dx, Dz, Ntw, Nsg, LEN, SHF, SVF,
      plus data rows for:
         SGM-Data, TELE-Data, TRIL-Data, LEVEL-Data, GPS-Data, INSAR-Data, SURF-Data, OTHER-Data,
         SGM-PHImx, TELE-PHImx, TRIL-PHImx, LEVEL-PHImx, GPS-PHImx, INSAR-PHImx, SURF-PHImx, OTHER-PHImx,
         SGM-Rmin, TELE-Rmin, TRIL-Rmin, LEVEL-Rmin, GPS-Rmin, INSAR-Rmin, SURF-Rmin, OTHER-Rmin.
    """
    data = {}
    lines = [line.strip('% ').strip() for line in block.splitlines() if line.strip()]
    for line in lines:
        if line.startswith("Invs :"):
            # Look for different parameters by checking keywords.
            if "Nx" in line and "Nz" in line:
                m_nx = re.search(r"Nx\s*=\s*([\d\.Ee\+\-]+)", line)
                m_nz = re.search(r"Nz\s*=\s*([\d\.Ee\+\-]+)", line)
                m_fmin = re.search(r"Fmin\s*=\s*([\d\.Ee\+\-]+)", line)
                m_fmax = re.search(r"Fmax\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_nx: data["Nx"] = float(m_nx.group(1))
                if m_nz: data["Nz"] = float(m_nz.group(1))
                if m_fmin: data["Fmin"] = float(m_fmin.group(1))
                if m_fmax: data["Fmax"] = float(m_fmax.group(1))
            elif "Dx" in line and "Dz" in line:
                m_dx = re.search(r"Dx\s*=\s*([\d\.Ee\+\-]+)", line)
                m_dz = re.search(r"Dz\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_dx: data["Dx"] = float(m_dx.group(1))
                if m_dz: data["Dz"] = float(m_dz.group(1))
            elif "Ntw" in line and "Nsg" in line:
                m_ntw = re.search(r"Ntw\s*=\s*([\d\.Ee\+\-]+)", line)
                m_nsg = re.search(r"Nsg\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_ntw: data["Ntw"] = float(m_ntw.group(1))
                if m_nsg: data["Nsg"] = float(m_nsg.group(1))
            elif "LEN" in line and "SHF" in line:
                m_len = re.search(r"LEN\s*=\s*([\d\.Ee\+\-]+)", line)
                m_shf = re.search(r"SHF\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_len: data["LEN"] = float(m_len.group(1))
                if m_shf: data["SHF"] = float(m_shf.group(1))
        elif line.startswith("SVF"):
            m = re.search(r"SVF\s*:\s*(\S+)", line)
            if m:
                data["SVF"] = m.group(1)
    # Process the "Data :" lines (we assume the first is header and the next is numeric)
    data_lines = [line for line in lines if line.startswith("Data :")]
    if len(data_lines) >= 2:
        header_line = data_lines[0]
        value_line = data_lines[1]
        headers = header_line.replace("Data :", "").strip().split()
        values = value_line.replace("Data :", "").strip().split()
        for h, v in zip(headers, values):
            try:
                data[f"{h}-Data"] = float(v)
            except:
                data[f"{h}-Data"] = v
    # Process PHImx (assume one line, with order SGM, TELE, TRIL, LEVEL, GPS, INSAR, SURF, OTHER)
    phimx_lines = [line for line in lines if line.startswith("PHImx:")]
    if phimx_lines:
        phimx_line = phimx_lines[0]
        parts = phimx_line.replace("PHImx:", "").strip().split()
        headers = ["SGM", "TELE", "TRIL", "LEVEL", "GPS", "INSAR", "SURF", "OTHER"]
        for h, v in zip(headers, parts):
            try:
                data[f"{h}-PHImx"] = float(v)
            except:
                data[f"{h}-PHImx"] = v
    # Process Rmin
    rmin_lines = [line for line in lines if line.startswith("Rmin :")]
    if rmin_lines:
        rmin_line = rmin_lines[0]
        parts = rmin_line.replace("Rmin :", "").strip().split()
        headers = ["SGM", "TELE", "TRIL", "LEVEL", "GPS", "INSAR", "SURF", "OTHER"]
        for h, v in zip(headers, parts):
            try:
                data[f"{h}-Rmin"] = float(v)
            except:
                data[f"{h}-Rmin"] = v
    return data

def parse_velocity_density(block):
    """
    Parses the velocity-density structure section.
    Expected:
      - A header line indicating the number of layers (e.g. "No.  of layers =  5")
      - A table with columns: DEPTH, P-VEL, S-VEL, DENS, QP, QS
      - The parsed rows will include a column 'number_of_layers' with the value from the header.
    Returns a list of dictionaries (one per row).
    """
    num_layers = None
    rows = []
    lines = [line.strip('% ').strip() for line in block.splitlines() if line.strip()]
    for line in lines:
        if "of layers" in line:
            m = re.search(r"of layers\s*=\s*(\d+)", line)
            if m:
                num_layers = int(m.group(1))
    for line in lines:
        # Look for rows that start with a digit (skip header text)
        if re.match(r"^\d", line):
            parts = line.split()
            if len(parts) >= 6:
                try:
                    row = {
                        "DEPTH": float(parts[0]),
                        "P-VEL": float(parts[1]),
                        "S-VEL": float(parts[2]),
                        "DENS": float(parts[3]),
                        "QP": float(parts[4]),
                        "QS": float(parts[5]),
                        "number_of_layers": num_layers
                    }
                    rows.append(row)
                except:
                    pass
    return rows

def parse_source_model(block):
    """
    Parses the source model parameters section.
    Expected:
      - A header line (with "LAT       LON       X==EW       Y==NS       Z       SLIP       TRUP")
      - Followed by rows of numeric data.
    Returns a list of dictionaries.
    """
    rows = []
    lines = block.splitlines()
    header_found = False
    for line in lines:
        # Clean up line by removing '%' and '-' characters.
        clean_line = line.strip('%- ').strip()
        if not clean_line:
            continue
        if not header_found:
            if "LAT" in clean_line and "LON" in clean_line:
                header_found = True
            continue
        parts = clean_line.split()
        if len(parts) >= 7:
            try:
                row = {
                    "LAT": float(parts[0]),
                    "LON": float(parts[1]),
                    "X==EW": float(parts[2]),
                    "Y==NS": float(parts[3]),
                    "Z": float(parts[4]),
                    "SLIP": float(parts[5]),
                    "TRUP": float(parts[6])
                }
                rows.append(row)
            except:
                pass
    return rows

def process_fsp_files(folder_path, output_folder):
    """
    For each .fsp file in folder_path:
      - Reads the file.
      - Segments its content into four parts:
          (finite-source, inversion-related parameters,
           velocity-density structure, and source model parameters)
      - Parses each part and appends the parsed data (with a 'filename' key) to lists.
    Finally, writes four CSV files (one per section) into output_folder.
    """
    finite_data = []      # One dictionary per file (finite-source rupture model)
    inversion_data = []   # One dictionary per file (inversion parameters)
    velocity_data = []    # List of rows (each a dictionary) from the velocity-density table
    source_model_data = []  # List of rows from the source model table

    fsp_files = glob.glob(os.path.join(folder_path, "*.fsp"))
    for fsp_file in fsp_files:
        with open(fsp_file, 'r', encoding='utf-8') as f:
            content = f.read()
        # Segment the file by scanning for section titles.
        sections = {"finite": [], "inversion": [], "velocity": [], "source": []}
        current_section = None
        for line in content.splitlines():
            if "FINITE-SOURCE RUPTURE MODEL" in line:
                current_section = "finite"
                continue
            elif "inversion-related parameters" in line:
                current_section = "inversion"
                continue
            elif "VELOCITY-DENSITY STRUCTURE" in line:
                current_section = "velocity"
                continue
            elif "SOURCE MODEL PARAMETERS" in line:
                current_section = "source"
                continue
            if current_section:
                sections[current_section].append(line)
        finite_block = "\n".join(sections["finite"])
        inversion_block = "\n".join(sections["inversion"])
        velocity_block = "\n".join(sections["velocity"])
        source_model_block = "\n".join(sections["source"])

        filename = os.path.basename(fsp_file)
        if finite_block.strip():
            parsed = parse_finite_source(finite_block)
            parsed["filename"] = filename
            finite_data.append(parsed)
        if inversion_block.strip():
            parsed = parse_inversion(inversion_block)
            parsed["filename"] = filename
            inversion_data.append(parsed)
        if velocity_block.strip():
            rows = parse_velocity_density(velocity_block)
            for row in rows:
                row["filename"] = filename
            velocity_data.extend(rows)
        if source_model_block.strip():
            rows = parse_source_model(source_model_block)
            for row in rows:
                row["filename"] = filename
            source_model_data.extend(rows)

    os.makedirs(output_folder, exist_ok=True)

    # Write finite-source CSV with the following columns:
    # Event, EventTAG, LAT, LON, DEP, LEN, WID, Mw, Mo, STRK, DIP, RAKE, Htop, HypX, HypZ, avTr, avVr
    finite_fixed = ["filename", "Event", "EventTAG", "LAT", "LON", "DEP", "LEN", "WID",
                    "Mw", "Mo", "STRK", "DIP", "RAKE", "Htop", "HypX", "HypZ", "avTr", "avVr"]
    finite_fields = get_fieldnames(finite_data, finite_fixed)
    with open(os.path.join(output_folder, "finite_source.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=finite_fields)
        writer.writeheader()
        for row in finite_data:
            # Fill missing fields with NaN
            for field in finite_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

    # Write inversion parameters CSV with the following columns:
    # Nx, Nz, Fmin, Fmax, Dx, Dz, Ntw, Nsg, LEN, SHF, SVF,
    # SGM-Data, TELE-Data, TRIL-Data, LEVEL-Data, GPS-Data, INSAR-Data, SURF-Data, OTHER-Data,
    # SGM-PHImx, TELE-PHImx, TRIL-PHImx, LEVEL-PHImx, GPS-PHImx, INSAR-PHImx, SURF-PHImx, OTHER-PHImx,
    # SGM-Rmin, TELE-Rmin, TRIL-Rmin, LEVEL-Rmin, GPS-Rmin, INSAR-Rmin, SURF-Rmin, OTHER-Rmin
    inversion_fixed = ["filename", "Nx", "Nz", "Fmin", "Fmax", "Dx", "Dz", "Ntw", "Nsg",
                       "LEN", "SHF", "SVF",
                       "SGM-Data", "TELE-Data", "TRIL-Data", "LEVEL-Data", "GPS-Data", "INSAR-Data", "SURF-Data", "OTHER-Data",
                       "SGM-PHImx", "TELE-PHImx", "TRIL-PHImx", "LEVEL-PHImx", "GPS-PHImx", "INSAR-PHImx", "SURF-PHImx", "OTHER-PHImx",
                       "SGM-Rmin", "TELE-Rmin", "TRIL-Rmin", "LEVEL-Rmin", "GPS-Rmin", "INSAR-Rmin", "SURF-Rmin", "OTHER-Rmin"]
    inversion_fields = get_fieldnames(inversion_data, inversion_fixed)
    with open(os.path.join(output_folder, "inversion_parameters.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=inversion_fields)
        writer.writeheader()
        for row in inversion_data:
            for field in inversion_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

    # Write velocity-density CSV with the following columns:
    # DEPTH, P-VEL, S-VEL, DENS, QP, QS, number_of_layers
    velocity_fixed = ["filename", "DEPTH", "P-VEL", "S-VEL", "DENS", "QP", "QS", "number_of_layers"]
    velocity_fields = get_fieldnames(velocity_data, velocity_fixed)
    with open(os.path.join(output_folder, "velocity_density.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=velocity_fields)
        writer.writeheader()
        for row in velocity_data:
            for field in velocity_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

    # Write source model CSV with the following columns:
    # LAT, LON, X==EW, Y==NS, Z, SLIP, TRUP
    source_fixed = ["filename","LAT", "LON", "X==EW", "Y==NS", "Z", "SLIP", "TRUP"]
    source_fields = get_fieldnames(source_model_data, source_fixed)
    with open(os.path.join(output_folder, "source_model.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=source_fields)
        writer.writeheader()
        for row in source_model_data:
            for field in source_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

if __name__ == '__main__':
    # Use raw strings or double backslashes in your Windows paths.
    folder_path = r"E:\SDA_CE\srcmod2024-12-02FSP\non-multisegment"  # Folder containing .fsp files
    output_folder = r"E:\SDA_CE\Dataset\non-multisegment"             # Folder where CSV files will be saved
    process_fsp_files(folder_path, output_folder)
    print("CSV files have been generated.")
```

```python
import os
import glob
import re
import csv

def get_fieldnames(data_list, fixed_columns):
    """
    Given a list of dictionaries and a list of fixed_columns,
    returns a list of field names with fixed_columns first and then
    any additional keys (sorted). Missing values will later be set as NaN.
    """
    union_keys = set()
    for row in data_list:
        union_keys |= set(row.keys())
    additional = sorted(list(union_keys - set(fixed_columns)))
    return fixed_columns + additional

def parse_finite_source(block):
    """
    Parses the finite-source rupture model section.
    Expected fields include:
      Event, EventTAG, LAT, LON, DEP, LEN, WID, Mw, Mo,
      STRK, DIP, RAKE, Htop, HypX, HypZ, avTr, avVr
    """
    data = {}
    lines = [line.strip('% ').strip() for line in block.splitlines() if line.strip()]
    for line in lines:
        if line.startswith("Event :"):
            m = re.search(r"Event\s*:\s*([^\[]+)", line)
            if m:
                event = m.group(1).strip()
                # Remove any date (dd/mm/yyyy)
                event = re.sub(r"\d{1,2}/\d{1,2}/\d{4}", "", event).strip()
                data["Event"] = event
        elif line.startswith("EventTAG:"):
            m = re.search(r"EventTAG:\s*(\S+)", line)
            if m:
                data["EventTAG"] = m.group(1)
        elif line.startswith("Loc  :"):
            m_lat = re.search(r"LAT\s*=\s*([\d\.\-]+)", line)
            m_lon = re.search(r"LON\s*=\s*([\d\.\-]+)", line)
            m_dep = re.search(r"DEP\s*=\s*([\d\.\-]+)", line)
            if m_lat:
                data["LAT"] = float(m_lat.group(1))
            if m_lon:
                data["LON"] = float(m_lon.group(1))
            if m_dep:
                data["DEP"] = float(m_dep.group(1))
        elif line.startswith("Size :"):
            m_len = re.search(r"LEN\s*=\s*([\d\.Ee\+\-]+)", line)
            m_wid = re.search(r"WID\s*=\s*([\d\.Ee\+\-]+)", line)
            m_mw = re.search(r"Mw\s*=\s*([\d\.Ee\+\-]+)", line)
            m_mo = re.search(r"Mo\s*=\s*([\d\.Ee\+\-]+)", line)
            if m_len:
                data["LEN"] = float(m_len.group(1))
            if m_wid:
                data["WID"] = float(m_wid.group(1))
            if m_mw:
                data["Mw"] = float(m_mw.group(1))
            if m_mo:
                data["Mo"] = float(m_mo.group(1))
        elif line.startswith("Mech :"):
            m_strk = re.search(r"STRK\s*=\s*([\d\.Ee\+\-]+)", line)
            m_dip = re.search(r"DIP\s*=\s*([\d\.Ee\+\-]+)", line)
            m_rake = re.search(r"RAKE\s*=\s*([\d\.Ee\+\-]+)", line)
            m_htop = re.search(r"Htop\s*=\s*([\d\.Ee\+\-]+)", line)
            if m_strk:
                data["STRK"] = float(m_strk.group(1))
            if m_dip:
                data["DIP"] = float(m_dip.group(1))
            if m_rake:
                data["RAKE"] = float(m_rake.group(1))
            if m_htop:
                data["Htop"] = float(m_htop.group(1))
        elif line.startswith("Rupt :"):
            m_hypx = re.search(r"HypX\s*=\s*([\d\.Ee\+\-]+)", line)
            m_hypz = re.search(r"HypZ\s*=\s*([\d\.Ee\+\-]+)", line)
            m_avtr = re.search(r"avTr\s*=\s*([\d\.Ee\+\-]+)", line)
            m_avvr = re.search(r"avVr\s*=\s*([\d\.Ee\+\-]+)", line)
            if m_hypx:
                data["HypX"] = float(m_hypx.group(1))
            if m_hypz:
                data["HypZ"] = float(m_hypz.group(1))
            if m_avtr:
                data["avTr"] = float(m_avtr.group(1))
            if m_avvr:
                data["avVr"] = float(m_avvr.group(1))
    return data

def parse_inversion(block):
    """
    Parses the inversion-related parameters section.
    Expected fields include:
      Nx, Nz, Fmin, Fmax, Dx, Dz, Ntw, Nsg, LEN, SHF, SVF,
      plus data rows (with keys like SGM-Data, TELE-Data, etc.),
      PHImx values and Rmin values.
    Additional fields (e.g. hr-GPS-Data, hr-GPS-PHImx, hr-GPS-Rmin)
    may be added if present.
    """
    data = {}
    lines = [line.strip('% ').strip() for line in block.splitlines() if line.strip()]
    for line in lines:
        if line.startswith("Invs :"):
            if "Nx" in line and "Nz" in line:
                m_nx = re.search(r"Nx\s*=\s*([\d\.Ee\+\-]+)", line)
                m_nz = re.search(r"Nz\s*=\s*([\d\.Ee\+\-]+)", line)
                m_fmin = re.search(r"Fmin\s*=\s*([\d\.Ee\+\-]+)", line)
                m_fmax = re.search(r"Fmax\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_nx:
                    data["Nx"] = float(m_nx.group(1))
                if m_nz:
                    data["Nz"] = float(m_nz.group(1))
                if m_fmin:
                    data["Fmin"] = float(m_fmin.group(1))
                if m_fmax:
                    data["Fmax"] = float(m_fmax.group(1))
            elif "Dx" in line and "Dz" in line:
                m_dx = re.search(r"Dx\s*=\s*([\d\.Ee\+\-]+)", line)
                m_dz = re.search(r"Dz\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_dx:
                    data["Dx"] = float(m_dx.group(1))
                if m_dz:
                    data["Dz"] = float(m_dz.group(1))
            elif "Ntw" in line and "Nsg" in line:
                m_ntw = re.search(r"Ntw\s*=\s*([\d\.Ee\+\-]+)", line)
                m_nsg = re.search(r"Nsg\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_ntw:
                    data["Ntw"] = float(m_ntw.group(1))
                if m_nsg:
                    data["Nsg"] = float(m_nsg.group(1))
            elif "LEN" in line and "SHF" in line:
                m_len = re.search(r"LEN\s*=\s*([\d\.Ee\+\-]+)", line)
                m_shf = re.search(r"SHF\s*=\s*([\d\.Ee\+\-]+)", line)
                if m_len:
                    data["LEN"] = float(m_len.group(1))
                if m_shf:
                    data["SHF"] = float(m_shf.group(1))
        elif line.startswith("SVF"):
            m = re.search(r"SVF\s*:\s*(\S+)", line)
            if m:
                data["SVF"] = m.group(1)
    # Process the "Data :" lines (header and values)
    data_lines = [line for line in lines if line.startswith("Data :")]
    if len(data_lines) >= 2:
        header_line = data_lines[0]
        value_line = data_lines[1]
        headers = header_line.replace("Data :", "").strip().split()
        values = value_line.replace("Data :", "").strip().split()
        for h, v in zip(headers, values):
            try:
                data[f"{h}-Data"] = float(v)
            except:
                data[f"{h}-Data"] = v
    # Process PHImx
    phimx_lines = [line for line in lines if line.startswith("PHImx:")]
    if phimx_lines:
        phimx_line = phimx_lines[0]
        parts = phimx_line.replace("PHImx:", "").strip().split()
        headers = ["SGM", "TELE", "TRIL", "LEVEL", "GPS", "INSAR", "SURF", "OTHER"]
        for h, v in zip(headers, parts):
            try:
                data[f"{h}-PHImx"] = float(v)
            except:
                data[f"{h}-PHImx"] = v
    # Process Rmin
    rmin_lines = [line for line in lines if line.startswith("Rmin :")]
    if rmin_lines:
        rmin_line = rmin_lines[0]
        parts = rmin_line.replace("Rmin :", "").strip().split()
        headers = ["SGM", "TELE", "TRIL", "LEVEL", "GPS", "INSAR", "SURF", "OTHER"]
        for h, v in zip(headers, parts):
            try:
                data[f"{h}-Rmin"] = float(v)
            except:
                data[f"{h}-Rmin"] = v
    return data

def parse_velocity_density(block):
    """
    Parses the velocity-density structure section.
    Expected header (number of layers) and then rows with:
      DEPTH, P-VEL, S-VEL, DENS, QP, QS.
    A column 'number_of_layers' is added based on the header.
    """
    num_layers = None
    rows = []
    lines = [line.strip('% ').strip() for line in block.splitlines() if line.strip()]
    for line in lines:
        if "of layers" in line:
            m = re.search(r"of layers\s*=\s*(\d+)", line)
            if m:
                num_layers = int(m.group(1))
    for line in lines:
        # Look for rows that start with a digit
        if re.match(r"^\d", line):
            parts = line.split()
            if len(parts) >= 6:
                try:
                    row = {
                        "DEPTH": float(parts[0]),
                        "P-VEL": float(parts[1]),
                        "S-VEL": float(parts[2]),
                        "DENS": float(parts[3]),
                        "QP": float(parts[4]),
                        "QS": float(parts[5]),
                        "number_of_layers": num_layers
                    }
                    rows.append(row)
                except:
                    pass
    return rows

def parse_multisegment_model(block):
    """
    Parses the multisegment model section.
    For each segment header it extracts:
      - SEGMENT number, STRIKE, DIP
      - LEN and WID,
      - Z2top (depth to top)
      - top_center_LAT and top_center_LON,
      - Hypocenter info: Along-strike (X) and down-dip (Z),
      - Nsbfs (number of subfaults)
    Then it reads the subfault table that follows (expected columns:
      LAT, LON, X==EW, Y==NS, Z, SLIP, and optionally TRUP).
    Each subfault row is combined with the segment-level info.
    """
    rows = []
    lines = [line.strip() for line in block.splitlines() if line.strip()]
    i = 0
    while i < len(lines):
        line = lines[i]
        # Look for a segment header line containing "SEGMENT" and "#"
        if "SEGMENT" in line and "#" in line:
            seg_info = {}
            # Extract SEGMENT number, STRIKE and DIP
            m_seg = re.search(r"SEGMENT\s*#\s*(\d+):.*STRIKE\s*=\s*([\d\.]+).*DIP\s*=\s*([\d\.]+)", line)
            if m_seg:
                seg_info["SEGMENT"] = int(m_seg.group(1))
                seg_info["STRIKE"] = float(m_seg.group(2))
                seg_info["DIP"] = float(m_seg.group(3))
            # Next line: LEN and WID
            i += 1
            if i < len(lines):
                m_lenwid = re.search(r"LEN\s*=\s*([\d\.]+).*WID\s*=\s*([\d\.]+)", lines[i])
                if m_lenwid:
                    seg_info["LEN"] = float(m_lenwid.group(1))
                    seg_info["WID"] = float(m_lenwid.group(2))
            # Next line: depth to top (Z2top)
            i += 1
            if i < len(lines):
                m_z2top = re.search(r"Z2top\s*=\s*([\d\.]+)", lines[i])
                if m_z2top:
                    seg_info["Z2top"] = float(m_z2top.group(1))
            # Skip lines until the top-center coordinates
            while i < len(lines) and "top-center" not in lines[i]:
                i += 1
            # Next line should include top-center coordinates
            if i < len(lines):
                i += 1
                if i < len(lines):
                    m_top = re.search(r"LAT\s*=\s*([\d\.]+).*LON\s*=\s*([-.\d]+)", lines[i])
                    if m_top:
                        seg_info["top_center_LAT"] = float(m_top.group(1))
                        seg_info["top_center_LON"] = float(m_top.group(2))
            # Next, look for the hypocenter info line (along-strike and down-dip)
            i += 1
            if i < len(lines):
                m_hypo = re.search(r"along-strike\s*\(X\)\s*=\s*([\d\.]+).*down-dip\s*\(Z\)\s*=\s*([\d\.]+)", lines[i])
                if m_hypo:
                    seg_info["Along-strike(X)"] = float(m_hypo.group(1))
                    seg_info["down-dip(Z)"] = float(m_hypo.group(2))
            # Next, get number of subfaults (Nsbfs)
            i += 1
            if i < len(lines):
                m_nsbfs = re.search(r"Nsbfs\s*=\s*(\d+)", lines[i])
                if m_nsbfs:
                    seg_info["Nsbfs"] = int(m_nsbfs.group(1))
            # Now skip until we find the subfault table header (a line containing "LAT" and "LON")
            while i < len(lines) and not (("LAT" in lines[i] and "LON" in lines[i]) or re.match(r"^[\d\.\-]", lines[i])):
                i += 1
            # If a header line is found, skip it
            if i < len(lines) and "LAT" in lines[i] and "LON" in lines[i]:
                i += 1
            # Now, read the subfault rows until a new segment header is encountered or the section ends.
            while i < len(lines) and lines[i] and not ("SEGMENT" in lines[i] and "#" in lines[i]):
                # Skip separator lines (e.g. lines of dashes)
                if set(lines[i]) <= set("- "):
                    i += 1
                    continue
                if re.match(r"^[\d\.\-]", lines[i]):
                    parts = lines[i].split()
                    subfault = {}
                    # Expecting at least 6 values: LAT, LON, X==EW, Y==NS, Z, SLIP
                    try:
                        subfault["subfault_LAT"] = float(parts[0])
                        subfault["subfault_LON"] = float(parts[1])
                        subfault["X==EW"] = float(parts[2])
                        subfault["Y==NS"] = float(parts[3])
                        subfault["Z"] = float(parts[4])
                        if len(parts) >= 6:
                            subfault["SLIP"] = float(parts[5])
                        if len(parts) >= 7:
                            subfault["TRUP"] = float(parts[6])
                        else:
                            subfault["TRUP"] = "NaN"
                    except:
                        pass
                    # Combine segment-level info with subfault row
                    combined = {**seg_info, **subfault}
                    rows.append(combined)
                    i += 1
                else:
                    i += 1
            continue
        else:
            i += 1
    return rows

def process_fsp_files(folder_path, output_folder):
    """
    For each .fsp file in folder_path:
      - Reads the file.
      - Segments its content into four parts:
          (finite-source rupture model,
           inversion-related parameters,
           velocity-density structure,
           multisegment model)
      - Parses each part and appends the parsed data (with a 'filename' key) to lists.
    Finally, writes four CSV files into output_folder.
    """
    finite_data = []      # Data from finite-source rupture model section
    inversion_data = []   # Data from inversion-related parameters section
    velocity_data = []    # Rows from velocity-density structure section
    multiseg_data = []    # Rows from multisegment model section

    fsp_files = glob.glob(os.path.join(folder_path, "*.fsp"))
    for fsp_file in fsp_files:
        with open(fsp_file, 'r', encoding='utf-8') as f:
            content = f.read()
        # Segment the file by scanning for section titles.
        sections = {"finite": [], "inversion": [], "velocity": [], "multiseg": []}
        current_section = None
        for line in content.splitlines():
            if "FINITE-SOURCE RUPTURE MODEL" in line:
                current_section = "finite"
                continue
            elif "inversion-related parameters" in line:
                current_section = "inversion"
                continue
            elif "VELOCITY-DENSITY STRUCTURE" in line:
                current_section = "velocity"
                continue
            elif "MULTISEGMENT MODEL" in line:
                current_section = "multiseg"
                continue
            if current_section:
                sections[current_section].append(line)
        finite_block = "\n".join(sections["finite"])
        inversion_block = "\n".join(sections["inversion"])
        velocity_block = "\n".join(sections["velocity"])
        multiseg_block = "\n".join(sections["multiseg"])

        filename = os.path.basename(fsp_file)
        if finite_block.strip():
            parsed = parse_finite_source(finite_block)
            parsed["filename"] = filename
            finite_data.append(parsed)
        if inversion_block.strip():
            parsed = parse_inversion(inversion_block)
            parsed["filename"] = filename
            inversion_data.append(parsed)
        if velocity_block.strip():
            rows = parse_velocity_density(velocity_block)
            for row in rows:
                row["filename"] = filename
            velocity_data.extend(rows)
        if multiseg_block.strip():
            rows = parse_multisegment_model(multiseg_block)
            for row in rows:
                row["filename"] = filename
            multiseg_data.extend(rows)

    os.makedirs(output_folder, exist_ok=True)

    # Write CSV 1: finite-source rupture model
    finite_fixed = ["filename", "Event", "EventTAG", "LAT", "LON", "DEP", "LEN", "WID",
                    "Mw", "Mo", "STRK", "DIP", "RAKE", "Htop", "HypX", "HypZ", "avTr", "avVr"]
    finite_fields = get_fieldnames(finite_data, finite_fixed)
    with open(os.path.join(output_folder, "finite_source.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=finite_fields)
        writer.writeheader()
        for row in finite_data:
            for field in finite_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

    # Write CSV 2: inversion-related parameters
    inversion_fixed = [
        "filename", "Nx", "Nz", "Fmin", "Fmax", "Dx", "Dz", "Ntw", "Nsg", "LEN", "SHF", "SVF",
        "SGM-Data", "TELE-Data", "TRIL-Data", "LEVEL-Data", "GPS-Data", "INSAR-Data", "SURF-Data", "OTHER-Data",
        "SGM-PHImx", "TELE-PHImx", "TRIL-PHImx", "LEVEL-PHImx", "GPS-PHImx", "INSAR-PHImx", "SURF-PHImx", "OTHER-PHImx",
        "SGM-Rmin", "TELE-Rmin", "TRIL-Rmin", "LEVEL-Rmin", "GPS-Rmin", "INSAR-Rmin", "SURF-Rmin", "OTHER-Rmin",
        "hr-GPS-Data", "hr-GPS-PHImx", "hr-GPS-Rmin"
    ]
    inversion_fields = get_fieldnames(inversion_data, inversion_fixed)
    with open(os.path.join(output_folder, "inversion_parameters.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=inversion_fields)
        writer.writeheader()
        for row in inversion_data:
            for field in inversion_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

    # Write CSV 3: velocity-density structure
    velocity_fixed = ["filename", "DEPTH", "P-VEL", "S-VEL", "DENS", "QP", "QS", "number_of_layers"]
    velocity_fields = get_fieldnames(velocity_data, velocity_fixed)
    with open(os.path.join(output_folder, "velocity_density.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=velocity_fields)
        writer.writeheader()
        for row in velocity_data:
            for field in velocity_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

    # Write CSV 4: multisegment model
    # The columns are as requested. Note that the subfault table’s LAT and LON are stored as 'subfault_LAT' and 'subfault_LON'
    multiseg_fixed = [
        "filename", "SEGMENT", "STRIKE", "DIP", "LEN", "WID", "Z2top",
        "top_center_LAT", "top_center_LON", "Along-strike(X)", "down-dip(Z)",
        "Nsbfs", "subfault_LAT", "subfault_LON", "X==EW", "Y==NS", "Z", "SLIP", "TRUP"
    ]
    multiseg_fields = get_fieldnames(multiseg_data, multiseg_fixed)
    with open(os.path.join(output_folder, "multisegment_model.csv"), "w", newline='', encoding='utf-8') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=multiseg_fields)
        writer.writeheader()
        for row in multiseg_data:
            for field in multiseg_fields:
                if field not in row:
                    row[field] = "NaN"
            writer.writerow(row)

if __name__ == '__main__':
    # Change these paths as needed (use raw strings or double backslashes on Windows)
    folder_path = r"E:\SDA_CE\srcmod2024-12-02FSP\multisegment"   # Folder containing .fsp files
    output_folder = r"E:\SDA_CE\Dataset\multisegment"      # Folder where CSV files will be saved
    process_fsp_files(folder_path, output_folder)
    print("CSV files have been generated.")
```
