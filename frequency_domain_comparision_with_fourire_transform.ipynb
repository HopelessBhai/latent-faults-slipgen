```python
import os
import numpy as np
from PIL import Image

# --- Configuration ---
# Path to your folder with original images
input_dir = "./Dataset/filtered_images_test" 
# Path to save the processed 2D arrays
output_dir = "./Dataset/original_test_npy_images" 
# Standard size for all images
target_size = (50, 50)

# Create output folder if not exists
os.makedirs(output_dir, exist_ok=True)

# Loop over all .png files
for file_name in os.listdir(input_dir):
    if file_name.endswith(".png"):
        # Read image
        img_path = os.path.join(input_dir, file_name)
        img = Image.open(img_path)
        
        # 1. Convert to Grayscale ("L" mode for Luminance)
        # This reduces 3 channels (R,G,B) to 1, creating a 2D array.
        img_gray = img.convert("L")
        
        # 2. Resize image to the standard size
        img_resized = img_gray.resize(target_size, Image.LANCZOS)
        
        # 3. Convert to numpy array
        img_array = np.array(img_resized, dtype=np.float32)
        
        # 4. Create the new filename
        new_file_name = file_name.replace("interpolated_slip_image_", "slip_array_")
        new_file_name = new_file_name.replace(".fsp.png", ".npy")

        # 5. Save the final 2D array as .npy
        npy_path = os.path.join(output_dir, new_file_name)
        np.save(npy_path, img_array)
        
        print(f"✅ Saved {npy_path} with shape {img_array.shape}")

print(f"🎉 All images converted to grayscale, resized to {target_size}, and saved as .npy!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft2, fftshift

def compare_slip_planes_fft(predicted_npy_path, original_npy_path):
    """
    Compares two slip plane images (.npy files) in the frequency domain using FFT.

    Args:
        predicted_npy_path (str): File path to the predicted slip plane .npy file.
        original_npy_path (str): File path to the original slip plane .npy file.
    """
    # 1. Load the slip plane images from .npy files
    predicted_slip = np.load(predicted_npy_path)
    original_slip = np.load(original_npy_path)

    # 2. Compute the 2D Fast Fourier Transform for both images
    # fftshift is used to move the zero-frequency component to the center of the spectrum
    fft_predicted = fftshift(fft2(predicted_slip))
    fft_original = fftshift(fft2(original_slip))

    # 3. Calculate the magnitude spectrum (log scale for better visualization)
    magnitude_predicted = np.log1p(np.abs(fft_predicted))
    magnitude_original = np.log1p(np.abs(fft_original))

    # 4. Visual Comparison: Plot the magnitude spectra
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    ax1 = axes[0]
    im1 = ax1.imshow(magnitude_predicted, cmap='viridis')
    ax1.set_title('Predicted Slip Plane - Magnitude Spectrum')
    ax1.set_xlabel('Frequency (X)')
    ax1.set_ylabel('Frequency (Y)')
    fig.colorbar(im1, ax=ax1, fraction=0.046, pad=0.04)

    ax2 = axes[1]
    im2 = ax2.imshow(magnitude_original, cmap='viridis')
    ax2.set_title('Original Slip Plane - Magnitude Spectrum')
    ax2.set_xlabel('Frequency (X)')
    ax2.set_ylabel('Frequency (Y)')
    fig.colorbar(im2, ax=ax2, fraction=0.046, pad=0.04)

    plt.suptitle('Frequency Domain Comparison of Slip Planes', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

    # 5. Quantitative Comparison: Normalized Cross-Correlation of magnitude spectra
    # This provides a single value representing the similarity of the frequency content.
    mag_pred_raw = np.abs(fft_predicted)
    mag_orig_raw = np.abs(fft_original)
    
    numerator = np.sum(mag_pred_raw * mag_orig_raw)
    denominator = np.sqrt(np.sum(mag_pred_raw**2)) * np.sqrt(np.sum(mag_orig_raw**2))
    
    cross_correlation = numerator / denominator
    
    print(f"Normalized Cross-Correlation of Magnitude Spectra: {cross_correlation:.4f}")
    print("A value closer to 1.0 indicates a higher similarity in frequency content.")


# --- How to use the function ---
# Replace with the actual paths to your .npy files
for files in os.listdir(folder_path_pred):
    if files.endswith('.npy'):
        predicted_file = os.path.join(folder_path_pred, files)
        original_file = os.path.join(folder_path, files)
        print(f"Comparing {files}...")
        compare_slip_planes_fft(predicted_file, original_file)
```

```python
import os
import numpy as np
from numpy.fft import fft2, fftshift
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def compare_and_plot(predicted_npy_path, original_npy_path):
    """
    Loads two .npy files, creates a comparison plot, and calculates cross-correlation.
    Returns the figure and the correlation score.
    """
    # Load the slip plane images from .npy files
    predicted_slip = np.load(predicted_npy_path)
    original_slip = np.load(original_npy_path)
    
    # Ensure arrays are 2D for FFT
    if predicted_slip.ndim > 2:
        predicted_slip = np.mean(predicted_slip, axis=2)
    if original_slip.ndim > 2:
        original_slip = np.mean(original_slip, axis=2)

    # Compute the 2D Fast Fourier Transform
    fft_predicted = fftshift(fft2(predicted_slip))
    fft_original = fftshift(fft2(original_slip))

    # Calculate the magnitude spectrum
    magnitude_predicted = np.log1p(np.abs(fft_predicted))
    magnitude_original = np.log1p(np.abs(fft_original))

    # --- Create the Plot ---
    fig, axes = plt.subplots(1, 2, figsize=(11, 5))
    
    # Plot for Predicted Slip Plane
    im1 = axes[0].imshow(magnitude_predicted, cmap='viridis')
    axes[0].set_title('Predicted Slip Plane - Magnitude Spectrum')
    axes[0].set_xlabel('Frequency (X)')
    axes[0].set_ylabel('Frequency (Y)')
    fig.colorbar(im1, ax=axes[0])

    # Plot for Original Slip Plane
    im2 = axes[1].imshow(magnitude_original, cmap='viridis')
    axes[1].set_title('Original Slip Plane - Magnitude Spectrum')
    axes[1].set_xlabel('Frequency (X)')
    axes[1].set_ylabel('Frequency (Y)')
    fig.colorbar(im2, ax=axes[1])

    fig.suptitle(f'Frequency Comparison: {os.path.basename(predicted_npy_path)}', fontsize=14)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    
    # --- Calculate Cross-Correlation ---
    mag_pred_raw = np.abs(fft_predicted)
    mag_orig_raw = np.abs(fft_original)
    
    numerator = np.sum(mag_pred_raw * mag_orig_raw)
    denominator = np.sqrt(np.sum(mag_pred_raw**2)) * np.sqrt(np.sum(mag_orig_raw**2))
    
    cross_correlation = 0.0 if denominator == 0 else numerator / denominator
    
    return fig, cross_correlation


def create_pdf_report(pred_folder, orig_folder, output_pdf_name):
    """
    Generates a multi-page PDF report with comparison plots and scores.
    """
    # Use a sorted list to ensure consistent ordering
    predicted_files = sorted([f for f in os.listdir(pred_folder) if f.endswith('.npy')])
    
    with PdfPages(output_pdf_name) as pdf:
        print(f"Generating PDF report: {output_pdf_name}")
        for filename in predicted_files:
            original_file_path = os.path.join(orig_folder, filename)
            
            if os.path.exists(original_file_path):
                predicted_file_path = os.path.join(pred_folder, filename)
                
                print(f"  - Processing: {filename}")
                
                # Generate the plot and get the score
                fig, score = compare_and_plot(predicted_file_path, original_file_path)
                
                # Save the plot to the PDF
                pdf.savefig(fig)
                plt.close(fig) # Close the figure to free up memory
                
                # Create a new page for the text information
                text_fig = plt.figure(figsize=(11, 2))
                text_fig.text(0.5, 0.6,
                              f"File: {filename}",
                              ha='center', va='center', fontsize=12, weight='bold')
                text_fig.text(0.5, 0.3,
                              f"Normalized Cross-Correlation of Magnitude Spectra: {score:.4f}",
                              ha='center', va='center', fontsize=12)
                plt.axis('off')
                
                # Save the text page to the PDF
                pdf.savefig(text_fig)
                plt.close(text_fig)
            else:
                print(f"  - Skipping: {filename} (no matching original file found)")

    print("🎉 PDF report generated successfully!")


# --- How to Use ---
# 1. Replace these folder paths with your actual directory paths
folder_path_orig = './Dataset/original_test_npy_images'
folder_path_pred = './Dataset/slip_arrays_inference'
output_pdf_file = "slip_plane_comparison_report.pdf"

# 2. Run the function
create_pdf_report(folder_path_pred, folder_path_orig, output_pdf_file)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.fft import fft2, fftshift
import os  # Added import for os

def compare_slip_planes_fft(predicted_npy_path, original_npy_path):
    """
    Compares two slip plane images (.npy files) in the frequency domain using FFT.

    Args:
        predicted_npy_path (str): File path to the predicted slip plane .npy file.
        original_npy_path (str): File path to the original slip plane .npy file.
    """
    # 1. Load the slip plane images from .npy files
    predicted_slip = np.load(predicted_npy_path)
    original_slip = np.load(original_npy_path)

    # 2. Compute the 2D Fast Fourier Transform for both images
    # fftshift is used to move the zero-frequency component to the center of the spectrum
    fft_predicted = fftshift(fft2(predicted_slip))
    fft_original = fftshift(fft2(original_slip))

    # 3. Calculate the magnitude spectrum (log scale for better visualization)
    magnitude_predicted = np.log1p(np.abs(fft_predicted))
    magnitude_original = np.log1p(np.abs(fft_original))

    # 4. Visual Comparison: Plot the magnitude spectra
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    ax1 = axes[0]
    im1 = ax1.imshow(magnitude_predicted, cmap='viridis')
    ax1.set_title('Predicted Slip Plane - Magnitude Spectrum')
    ax1.set_xlabel('Frequency (X)')
    ax1.set_ylabel('Frequency (Y)')
    fig.colorbar(im1, ax=ax1, fraction=0.046, pad=0.04)

    ax2 = axes[1]
    im2 = ax2.imshow(magnitude_original, cmap='viridis')
    ax2.set_title('Original Slip Plane - Magnitude Spectrum')
    ax2.set_xlabel('Frequency (X)')
    ax2.set_ylabel('Frequency (Y)')
    fig.colorbar(im2, ax=ax2, fraction=0.046, pad=0.04)

    plt.suptitle('Frequency Domain Comparison of Slip Planes', fontsize=16)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

    # 5. Quantitative Comparison: Mean Squared Error (MSE) of magnitude spectra
    # This provides a single value representing the difference in frequency content.
    mag_pred_raw = np.abs(fft_predicted)
    mag_orig_raw = np.abs(fft_original)
    
    # Calculate MSE
    mse = np.mean((mag_pred_raw - mag_orig_raw)**2)
    
    print(f"Mean Squared Error (MSE) of Magnitude Spectra: {mse:.4f}")
    print("A value closer to 0.0 indicates a higher similarity in frequency content (less error).")


# --- How to use the function ---
# Replace with the actual paths to your .npy files
folder_path = './Dataset/original_test_npy_images'
folder_path_pred = './Dataset/slip_arrays_inference'

# Check if paths exist before looping to avoid errors
if os.path.exists(folder_path_pred) and os.path.exists(folder_path):
    for files in os.listdir(folder_path_pred):
        if files.endswith('.npy'):
            predicted_file = os.path.join(folder_path_pred, files)
            original_file = os.path.join(folder_path, files)
            
            # Ensure the corresponding original file also exists
            if os.path.exists(original_file):
                print(f"Comparing {files}...")
                compare_slip_planes_fft(predicted_file, original_file)
            else:
                print(f"Warning: Original file not found for {files}")
else:
    print(f"Warning: One or both directories do not exist. Please check your paths:")
    print(f"Predicted path: {folder_path_pred}")
    print(f"Original path: {folder_path}")
```
